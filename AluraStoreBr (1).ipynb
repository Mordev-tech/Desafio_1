# 3. Média de Avaliação das Lojas

```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

```python
avaliacao_media = loja.groupby(["Produto"])["Avaliação da compra"].mean().reset_index().round(2)
avaliacao = avaliacao_media.sort_values(by="Avaliação da compra", ascending=False)

print(avaliacao)
```

# 4. Produtos Mais e Menos Vendidos

```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

```python
produtos_mais_vendidos = (loja.groupby("Produto").size().reset_index(name="Quantidade Vendida").sort_values(by="Quantidade Vendida", ascending=False)).head(10)

produtos_menos_vendidos = (loja.groupby("Produto").size().reset_index(name="Quantidade Vendida").sort_values(by="Quantidade Vendida", ascending=True)).head(10)

print("Resultado dos produtos mais vendidos: \n", produtos_mais_vendidos)
print("\n" + "-"*100 + "\n")
print("Resultado dos produtos menos vendidos: \n", produtos_menos_vendidos)
```
